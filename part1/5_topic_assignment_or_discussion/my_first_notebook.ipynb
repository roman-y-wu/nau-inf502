## Insertion Sort

I chose **insertion sort** as the sorting strategy for this notebook. Picking the right strategy matters because algorithms differ in time complexity, memory usage, and how they behave on different data distributions. Insertion sort is easy to implement, stable (equal elements keep their relative order), and works very well on small or nearly sorted datasets, so it is a solid first algorithm to learn.

The core idea is to view the array as a sorted region and an unsorted region. Starting from the second element, take the next unsorted value and compare it to the sorted region from right to left, shifting larger values one position to the right until the correct insertion point is found. Insert the value there and repeat until the entire array is in ascending order.


### Load Unsorted Data


```python
# Create an array with 15 unsorted numbers
unsorted_data = [42, 3, 15, 27, 8, 19, 1, 34, 12, 5, 23, 9, 16, 2, 30]
```

### Sorting Function


```python
def insertion_sort(arr):
    """Sort the array in ascending order using insertion sort and return a new array."""
    result = arr.copy()
    for i in range(1, len(result)):
        key = result[i]
        j = i - 1
        while j >= 0 and result[j] > key:
            result[j + 1] = result[j]
            j -= 1
        result[j + 1] = key
    return result
```

### Run the Sorting Function


```python
# (1) Print the unsorted array
print("Unsorted array:", unsorted_data)

# (2) Call the insertion sort function
sorted_data = insertion_sort(unsorted_data)

# (3) Print the sorted array
print("Sorted array:", sorted_data)
```
